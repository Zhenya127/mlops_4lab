```python
import pandas as pd
```

```python
titanic = pd.read_csv('datasets/titanic.csv', sep=',')
```

```python
titanic.head()
```

```python
import numpy as np
```

```python
titanic['younth'] = np.where(titanic['Age'] > 40, 'not young', 'young')
titanic.to_csv('datasets/titanic.csv', index= False )
titanic.sample(5)
```

```python
titanic["Age"] = titanic["Age"].fillna(titanic["Age"].mean())
titanic.to_csv("datasets/titanic.csv", sep=",")
titanic.sample(5)
```

```python
from sklearn.preprocessing import OneHotEncoder
one_hot_encoder = OneHotEncoder(drop = 'first')
data  = one_hot_encoder.fit_transform(titanic[['Sex']]).toarray()
column_names = one_hot_encoder.get_feature_names_out()
data_onehot = pd.DataFrame(data, columns=column_names)
```

```python
titanic = pd.concat(
    [titanic.reset_index(drop=True).drop(['Sex'], axis=1), data_onehot], 
    axis=1
)
```

```python
titanic.to_csv("datasets/titanic.csv", sep=",")
titanic.sample(5)
```
